```python
import numpy as np

def entropy(n: int = 5):
    prob = np.random.dirichlet(np.ones(n), size=1)
    return -np.sum(prob * np.log(prob))
```

```python
entropy(10)
```

```python
def convex_func(x):
    return x * np.log(x)
```

```python
import matplotlib.pyplot as plt
x = np.linspace(0.5, 1000, 100000)
y = convex_func(x)
plt.plot(x, y)
```
